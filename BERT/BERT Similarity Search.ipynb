### Importing Dependencies

```python
import numpy as np
import pandas as pd

import tensorflow as tf
from transformers import BertTokenizer, TFBertModel
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

### Loading the Dataset

```python
df = pd.read_csv("/content/FHR_NLP_Queries - Sheet1.csv")
```

```python
df.head()
```

### Global Variables

```python
checkpoint = "bert-base-nli-mean-tokens"
```

### Loading the model

```python
model = SentenceTransformer(checkpoint)
```

### Tokenization

```python
documents = model.encode(df["Natural Language "].tolist())
```

```python
documents
```

### Calculate pairwise cosine similarity between embeddings


```python
similarity_matrix = cosine_similarity([documents[0]], documents[1:])
```
