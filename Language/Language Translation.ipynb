### Imports

```python
from transformers import pipeline
```

### Creating a Text2TextGenerationPipeline for language translation

```python
pipe = pipeline(task='text2text-generation', model='facebook/m2m100_418M')
```

### Translating

```python
pipe("That is a flower", forced_bos_token_id=pipe.tokenizer.get_lang_id(lang='es'))
```

```python
pipe("Es una flor", forced_bos_token_id=pipe.tokenizer.get_lang_id(lang='en'))
```
